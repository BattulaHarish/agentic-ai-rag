## Ollama Embeddings

```python
from langchain_community.embeddings import OllamaEmbeddings
```

```python
embeddings=(
    OllamaEmbeddings(model="gemma:2b")
)
```

```python
r1=embeddings.embed_documents(
    [
        "Harish is a Good Boy",
        "Harish is a Bad Boy"
    ]
)
```

```python
len(r1[0])
```

```python
query_result=embeddings.embed_query("Who is Bad Boy")
```

```python
print(query_result)
```
