```python
import os
from dotenv import load_dotenv
load_dotenv() #load all the environment variables
```

```python
os.environ["OPENAI_API_KEY"]=os.getenv("OPENAI_API_KEY")
```

```python
from langchain_openai import OpenAIEmbeddings
embeddings=OpenAIEmbeddings(model="text-embedding-3-large")
embeddings
```

```python
text="this is sample project on Agentic Ai using RAG"
query_result=embeddings.embed_query(text)
query_result
```

```python
len(query_result)
```

```python
from langchain_community.document_loaders import TextLoader
loader=TextLoader('speech.txt')
docs=loader.load()
docs
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
text_splitter = RecursiveCharacterTextSplitter(chunk_size=500,chunk_overlap=50)
final_document=text_splitter.split_documents(docs)
final_document
```

```python
## vector Embeddings and vector storeDb

from langchain_community.vectorstores import Chroma

db=Chroma.from_documents(final_document,embeddings)
```

```python
query="The evolutionary origin of speech is subject to debate and speculation."
retrieved_result=db.similarity_search(query)
print(retrieved_result)
```
