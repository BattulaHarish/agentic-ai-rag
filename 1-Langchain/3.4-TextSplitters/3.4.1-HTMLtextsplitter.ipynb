### HTML text splitter

```python
from langchain_text_splitters import HTMLHeaderTextSplitter
html_string="""
<!doctype html>
<html lang="en">
<head><meta charset="utf-8"/><title>Article</title></head>
<body>
  <article>
    <header>
      <h1>The Rise of RAG Systems</h1>
      <p>By Harish • Nov 27, 2025</p>
    </header>

    <section>
      <h2>Introduction</h2>
      <p>Retrieval-augmented generation (RAG) combines retrieval with generation to produce accurate, grounded responses.</p>
    </section>

    <section>
      <h2>Architecture</h2>
      <h3>1. Ingestion</h3>
      <p>Documents are loaded, cleaned, and split into chunks.</p>
      <h3>2. Embedding</h3>
      <p>Chunks are embedded and stored in a vector DB.</p>
    </section>

    <footer>
      <p>Tags: AI, RAG, NLP</p>
    </footer>
  </article>
</body>
</html>
"""

headers_to_split_on = [
    ("h1", "header"),
    ("h2", "section"),
]

html_splitter = HTMLHeaderTextSplitter(headers_to_split_on)
html_header_split=html_splitter.split_text(html_string)
html_header_split

```

```python
from langchain_text_splitters import HTMLHeaderTextSplitter
html_string="""
<!DOCTYPE html>
<html>
<head>
    <title>AI Overview</title>
</head>
<body>

<h1>Artificial Intelligence</h1>
<p>Artificial Intelligence (AI) is the simulation of human intelligence processes by machines.</p>

<h2>Machine Learning</h2>
<p>Machine Learning is a subset of AI that allows systems to learn and improve from experience.</p>

<h3>Supervised Learning</h3>
<p>Supervised learning uses labeled data to train models.</p>

<h3>Unsupervised Learning</h3>
<p>Unsupervised learning discovers hidden patterns from unlabeled data.</p>

<h2>Deep Learning</h2>
<p>Deep Learning uses neural networks with many layers to analyze complex data.</p>

<h3>Convolutional Neural Networks</h3>
<p>CNNs are mainly used for image processing tasks.</p>

<h3>Recurrent Neural Networks</h3>
<p>RNNs are used for sequential data like text and speech.</p>

<h1>Applications of AI</h1>
<p>AI is used in healthcare, finance, robotics, gaming, and more.</p>

</body>
</html>
"""


headers_to_split_on = [ 
    ("h1", "Header 1"),
    ("h2", "Header 2"),
    ("h3", "Header 3"),
]

html_splitter=HTMLHeaderTextSplitter(headers_to_split_on)
html_header_split=html_splitter.split_text(html_string)
print(html_header_split)
```
